# Proyecto Final: Interconnect

### **Plan de Trabajo:** 



**Propósito del Proyecto:**
Pronosticar la tasa de cancelación de clientes en la empresa de telecomunicaciones Interconnect, con el fin de retenerlos mediante promociones y opciones de planes especiales.

**Objetivo Principal:**
Crear un modelo que pronostique los clientes en riesgo de abandonar alguno de los servicios.


---
#### Fase 1: Preprocesamiento y Evaluación de Datos

**Objetivo:**
Preprocesar y evaluar los datos disponibles para asegurar su calidad y adecuación para el modelado.

**Actividades:**
1. Tengo a disposición datos recopilados por el equipo de marketing en cuanto a la información de los clientes.
2. Limpiar los datos eliminando valores nulos, duplicados y outliers.
3. Realizar un análisis exploratorio de datos (EDA) para entender las características y patrones.
4. Normalizar y transformar los datos según sea necesario.
5. Dividir los datos en conjuntos de entrenamiento y prueba.

---
#### Fase 2: Determinación del Tipo de Tarea de Machine Learning

**Objetivo:**
Identificar la naturaleza del problema y seleccionar el tipo adecuado de tarea de machine learning.

**Actividades:**
1. Revisar los resultados del EDA.
2. Confirmar si el problema es de clasificación binaria (cliente se queda vs. cliente se va).


---
#### Fase 3: Desarrollo y Entrenamiento de Modelos

**Objetivo:**
Desarrollar y entrenar varios modelos para pronosticar la tasa de cancelación de clientes.

**Actividades:**
1. Seleccionar algoritmos de machine learning adecuados (Regresión Logística, Árboles de Decisión, Random Forest, XGBoost, etc.).
2. Entrenar múltiples modelos utilizando el conjunto de datos de entrenamiento.

---
#### Fase 4: Selección del Mejor Modelo

**Objetivo:**
Seleccionar el modelo con la métrica AUC_ROC más alta (≥0.88).

**Actividades:**
1. Evaluar cada modelo utilizando el conjunto de prueba.
2. Calcular la métrica AUC_ROC para cada modelo.
3. Seleccionar el modelo que cumpla con el criterio establecido (AUC_ROC ≥ 0.88).


---
#### Fase 5: Prueba del Modelo Seleccionado

**Objetivo:**
Probar el modelo seleccionado con un conjunto de datos de validación independiente.

**Actividades:**
1. Obtener un conjunto de datos de validación que no se haya usado en el entrenamiento.
2. Evaluar el rendimiento del modelo con este conjunto de datos.
3. Ajustar el modelo si es necesario basado en los resultados.

---
#### Fase 6: Preparación del Informe de Resultados

**Objetivo:**
Preparar un informe detallado de los resultados obtenidos y las recomendaciones.

**Actividades:**
1. Documentar el proceso seguido y los resultados obtenidos en cada fase.
2. Incluir gráficos y tablas relevantes que expliquen los hallazgos.
3. Proporcionar recomendaciones para la implementación de promociones y planes especiales basados en el modelo.
